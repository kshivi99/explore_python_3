# Advanced Strings

Always remmember - strings are immutable. So you cannot change it - you have to store the change in a new object.

## Case Change

```python
s = "hello world"
```

```python
# Capitalizes first alphabet

s.capitalize()
```

```python
# Capitalizes whole word

u_s = s.upper()
u_s
```

```python
# Lowercase whole word

u_s.lower()
```

## Location, Counting & Formatting

```python
s.count('o')
```

```python
# Index of first occurance

s.find('o')
```

```python
s.center(20, '-')
```
