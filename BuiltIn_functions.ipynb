## map()

map() in python takes two or more arguments. First is the function. Followed by list of arguments.
e.g. map(int, input().split()) will take all the inputs, split it with spaces, convert them to integer and return an iterable.

```python
def squ(num):
    return num * num

num = [1, 2, 3, 4]
l = list(map(squ, num))
print(l)
```

Lambda has a great use here in the map.

```python
lam = list(map((lambda x: x*x), num))
print(lam)
```

With multiple arguments

```python
a1 = [1, 2, 3]
a2 = [10, 10, 10]

print(list(map((lambda x,y:x+y), a1, a2)))
```

## zip()

zip() takes two or more iterables. It returns a iterable that is combination of these.

```python
a1 = [1, 2, 3]
a2 = [11, 12, 13]
a3 = [101, 102, 103]

list(zip(a1, a2, a3))
```

zip can be used to swap key-value of dictionary

```python
d1 = {'a':1, 'b':2, 'c':3}
dres = {}

for k,v in zip(d1, d1.values()):
    dres[v] = k
    
dres
```

## reduce()

reduce() takes two parameters : (function, sequence). The function takes two argument from the list, perform operation, returns a value, and then run the function again with the result and next element in the list till list goes empty

There are few simple applications i could think right off the bat - max_of_list. Let's implement and see.

```python
from functools import reduce

num = [23, 4, 28, 99, 1, 34]

max_func = lambda x,y: x if x>y else y
reduce(max_func, num)
```

## all() and any()

```python
#all - returns True if all elements of an iterable is True
#any - returns True if any one element is True

lst = [True, False, True]

print(all(lst))
print(any(lst))
```

## enumerate()

Enumerate allows you to keep a count as you iterate through an object. It does this by returning a tuple in the form (count,element).

```python
# Let's use it to create a dictionary

lst = ['a', 'b', 'c']
dic = {}

for k,v in enumerate(lst):
    dic[k] = v

dic
```

We can start with non zero value

```python
lst = ['Tues', 'Wed', 'Thurs']

for count,day in enumerate(lst, start=2):
    print(count, day)
```
